```python
import pandas as pd
```

```python
baselines = pd.read_csv('baselines.csv')
baselines.drop(columns=['moderation'], inplace=True)
baselines.columns = ['prompt', 'baseline_response']

immunised = pd.read_csv('immunized2.csv')
immunised.drop(columns=['prompt'], inplace=True)
immunised.columns = ['immnuised_response', 'immunised_moderation']


success_attack = pd.read_csv('success_attack2.csv')
success_attack.drop(columns=['prompt'], inplace=True)
success_attack.columns = ['non_immunised_response', 'non_immunised_moderation']
```

```python
merged_df = pd.concat([baselines, immunised, success_attack], axis=1)
```

```python
for record in merged_df.loc[(success_attack['non_immunised_moderation'] != 'safe') & \
                (immunised['immunised_moderation'] == 'safe')].iterrows():

    print('PROMPT:')
    print(record[1]['prompt'],'\n')
    print('BASELINE RESPONSE:')
    print(record[1]['baseline_response'], '\n')
    print('BASELINE RESPONSE UNDER ATTACK:')
    print(record[1]['non_immunised_response'], '\n')
    print('IMMUNISED RESPONSE UNDER ATTACK:')
    print(record[1]['immnuised_response'], '\n')
```
